```python
import numpy as np

# Calculate P/E Ratio
tesla['P/E Ratio'] = tesla['prccq'] / (tesla['niq'] / tesla['cshoq'])
tesla['P/E Ratio'] = tesla['P/E Ratio'].replace([np.inf, -np.inf], np.nan).fillna(0)

# Calculate Market Cap
tesla['Market Cap'] = tesla['prccq'] * tesla['cshoq']

# Calculate Total Debt
tesla['Total Debt'] = tesla['dlttq'] + tesla['dlcq']

# Calculate EV (Enterprise Value)
tesla['EV'] = tesla['Market Cap'] + tesla['Total Debt'] - tesla['cheq']

# Calculate EV/EBITDA
tesla['EV/EBITDA'] = tesla['EV'] / tesla['oibdpq']
tesla['EV/EBITDA'] = tesla['EV/EBITDA'].replace([np.inf, -np.inf], np.nan).fillna(0)

# Calculate P/S Ratio
tesla['P/S Ratio'] = tesla['Market Cap'] / tesla['saleq']
tesla['P/S Ratio'] = tesla['P/S Ratio'].replace([np.inf, -np.inf], np.nan).fillna(0)

# Calculate ROIC (Return on Invested Capital)
tesla['ROIC'] = tesla['niq'] / (tesla['dlttq'] + tesla['ceqq'])
tesla['ROIC'] = tesla['ROIC'].replace([np.inf, -np.inf], np.nan).fillna(0)

# Calculate Net Debt
tesla['Net Debt'] = tesla['Total Debt'] - tesla['cheq']

# Calculate Net Debt/EBITDA
tesla['Net Debt/EBITDA'] = tesla['Net Debt'] / tesla['oibdpq']
tesla['Net Debt/EBITDA'] = tesla['Net Debt/EBITDA'].replace([np.inf, -np.inf], np.nan).fillna(0)

# Calculate Interest Coverage
tesla['Interest Coverage'] = tesla['oibdpq'] / tesla['xintq']
tesla['Interest Coverage'] = tesla['Interest Coverage'].replace([np.inf, -np.inf], np.nan).fillna(0)

# Calculate FCF (Free Cash Flow)
tesla['FCF'] = tesla['oibdpq'] - tesla['capxy']

# Calculate FCF Yield
tesla['FCF Yield'] = tesla['FCF'] / tesla['Market Cap']
tesla['FCF Yield'] = tesla['FCF Yield'].replace([np.inf, -np.inf], np.nan).fillna(0)

print("Additional financial ratios calculated and added to the tesla DataFrame.")
```

## Prepare Ratios for Scoring

### Subtask:
Select the relevant financial ratios from the `financial_ratios_df` DataFrame, normalize them, and handle any missing values.


**Reasoning**:
To prepare the financial ratios for scoring, I will first define the list of relevant ratio columns. Then, I will create a new DataFrame `health_score_data` by selecting these columns from `financial_ratios_df`.



```python
ratio_columns = [
    'Gross Profit Margin',
    'Operating Profit Margin',
    'Net Profit Margin',
    'Current Ratio',
    'Debt-to-Equity Ratio',
    'P/E Ratio',
    'EV/EBITDA',
    'P/S Ratio',
    'ROIC',
    'Net Debt/EBITDA',
    'Interest Coverage',
    'FCF Yield'
]

health_score_data = financial_ratios_df[ratio_columns].copy()
print("Health score data DataFrame created with selected ratios.")
```

**Reasoning**:
To normalize the financial ratios and handle any remaining missing values, I will apply Min-Max scaling to each column in the `health_score_data` DataFrame and then fill any `NaN` values with 0.0. Finally, I will display the first few rows of the updated DataFrame to verify the changes.



```python
for col in health_score_data.columns:
    current_series = financial_ratios_df[col] # Use original financial_ratios_df for robust min/max/positive max
    min_val = current_series.min()
    max_val = current_series.max()

    if max_val - min_val == 0:
        health_score_data[col] = 0.0 # If all values are the same, normalized value is 0
    else:
        # Special handling for ratios where negative values are bad and lower/higher might be ideal
        if col in ['EV/EBITDA', 'Net Debt/EBITDA', 'Debt-to-Equity Ratio']: # Lower is better, but negative/high positive is bad
            max_positive_val_for_col = current_series[current_series > 0].max()
            if pd.isna(max_positive_val_for_col) or max_positive_val_for_col == 0: # If all values are negative or zero
                health_score_data[col] = 0.0 # Assign 0 score (bad)
            else:
                # Normalize positive values such that lower is better (0 -> 1, max_positive -> 0)
                # And negative/zero values explicitly map to 0 (bad)
                temp_scores = 1 - (current_series / max_positive_val_for_col)
                temp_scores[current_series <= 0] = 0.0 # Negative or zero values get 0 score
                health_score_data[col] = temp_scores

        elif col == 'P/E Ratio': # Higher is better, but negative is bad
            max_positive_val_for_col = current_series[current_series > 0].max()
            if pd.isna(max_positive_val_for_col) or max_positive_val_for_col == 0: # If all values are negative or zero
                health_score_data[col] = 0.0 # Assign 0 score (bad)
            else:
                # Normalize positive values such that higher is better (0 -> 0, max_positive -> 1)
                # And negative/zero values explicitly map to 0 (bad)
                temp_scores = current_series / max_positive_val_for_col
                temp_scores[current_series <= 0] = 0.0 # Negative or zero values get 0 score
                health_score_data[col] = temp_scores

        else: # Standard min-max normalization for other ratios
              # For these, 0 = min_val (worst), 1 = max_val (best)
            normalized_series = (current_series - min_val) / (max_val - min_val)

            # Invert if ratio_direction is 'lower is better'
            if ratio_directions.get(col) == -1:
                health_score_data[col] = 1 - normalized_series
            else: # Higher is better
                health_score_data[col] = normalized_series

health_score_data.fillna(0.0, inplace=True) # Final fillna for any remaining NaNs

print("Normalized financial ratios in 'health_score_data' with NaN values filled:")
print(health_score_data.head())
```

## Define Health Score Function

### Subtask:
Develop a Python function that takes the normalized financial ratios and a dictionary of weights as input. This function will calculate a composite health score based on the weighted sum of the ratios. The function should also consider the 'direction' of each ratio (e.g., higher P/E is usually better, but higher Debt-to-Equity is usually worse).


**Reasoning**:
To implement the `calculate_health_score` function, I will first define the `ratio_directions` dictionary to specify whether a higher value for each financial ratio is better (1) or worse (-1). Then, I will create the `calculate_health_score` function that iterates through the normalized ratios, adjusts them based on their direction, applies the provided weights, and sums them to produce a composite health score.



```python
ratio_directions = {
    'Gross Profit Margin': 1, # Higher is better
    'Operating Profit Margin': 1, # Higher is better
    'Net Profit Margin': 1, # Higher is better
    'Current Ratio': 1, # Higher is generally better (up to a point)
    'Debt-to-Equity Ratio': -1, # Lower is better (specific handling for negative in normalization)
    'P/E Ratio': 1, # Higher (positive) can indicate growth expectation, but negative is bad (specific handling in normalization)
    'EV/EBITDA': -1, # Lower is generally better, but negative is bad (specific handling in normalization)
    'P/S Ratio': 1, # Higher can indicate growth expectation
    'ROIC': 1, # Higher is better
    'Net Debt/EBITDA': -1, # Lower is better, but negative is bad (specific handling in normalization)
    'Interest Coverage': 1, # Higher is better
    'FCF Yield': 1 # Higher is better
}

def calculate_health_score(df, weights):
    score_series = pd.Series(0.0, index=df.index)
    for ratio in df.columns:
        if ratio in weights: # All normalization and direction handling is done before this function
            score_series += df[ratio] * weights[ratio]
    return score_series

print("Health score calculation function and ratio directions defined.")
```

Now that the `ratio_directions` dictionary, `calculate_health_score` function, and `ipywidgets` sliders for the default weights are initialized, the next step is to create the interactive interface. This involves defining an `update_plot` function that will take the slider values (new weights) as input, calculate the health score, and then generate a plot of the health score over time. Finally, the `interact` function from `ipywidgets` will be used to link the sliders to the `update_plot` function.



```python
def update_plot(**new_weights):
    current_weights = {ratio: new_weights[ratio] for ratio in ratio_columns}

    # Normalize weights so they sum to 1
    total_weight = sum(current_weights.values())
    if total_weight == 0:
        # Handle case where all weights are 0 to avoid division by zero
        normalized_weights = {ratio: 0 for ratio in ratio_columns}
    else:
        normalized_weights = {ratio: weight / total_weight for ratio, weight in current_weights.items()}

    # Calculate health score with normalized weights
    tesla['Health Score'] = calculate_health_score(health_score_data, normalized_weights)

    # Scale Health Score to 0-100 for better interpretation
    max_possible_score = len(ratio_columns) # Maximum score before scaling to 100
    tesla['Health Score (Percentage)'] = (tesla['Health Score'] / max_possible_score) * 100

    plt.figure(figsize=(12, 6))
    sns.lineplot(data=tesla, x=tesla.index, y='Health Score (Percentage)', marker='o', color='blue', label='Tesla Health Score')
    plt.title('Tesla Financial Health Score Over Time (Adjustable Weights)')
    plt.xlabel('Date')
    plt.ylabel('Health Score (Percentage)')
    plt.legend()
    plt.grid(True)
    plt.ylim(0, 100) # Set y-axis limit to 0-100 for percentage display

    # Add thresholds for 'Good' and 'Bad'
    plt.axhspan(0, 40, color='red', alpha=0.1, label='Bad (0-40%)')
    plt.axhspan(40, 70, color='yellow', alpha=0.1, label='Moderate (40-70%)')
    plt.axhspan(70, 100, color='green', alpha=0.1, label='Good (70-100%)')

    plt.show()

# Create the interactive interface
interactive_plot = widgets.interactive(update_plot, **weight_sliders)
display(interactive_plot)

print("Interactive health score tool created. Adjust the sliders to see the dynamic health score trend.")
```

## Prepare Tesla's Quarterly Ratios for Comparison

### Subtask:
Extract all quarterly financial ratios for Tesla from the `financial_ratios_df` DataFrame.


**Reasoning**:
To prepare Tesla's quarterly ratios for comparison with firm standards, I will create a new DataFrame `tesla_comparison_ratios` by selecting the specified financial ratio columns from the `financial_ratios_df`.



```python
tesla_comparison_ratios = financial_ratios_df[[
    'P/E Ratio',
    'P/S Ratio',
    'Net Profit Margin',
    'Operating Profit Margin',
    'Gross Profit Margin',
    'Debt-to-Equity Ratio',
    'Interest Coverage',
    'Current Ratio',
    'Net Debt/EBITDA',
    'ROIC'
]].copy()

print("Tesla's quarterly financial ratios for comparison extracted into 'tesla_comparison_ratios' DataFrame.")
print(tesla_comparison_ratios.head())
```

**Reasoning**:
To calculate the percentage difference between Tesla's ratios and the firm standards, I will iterate through each ratio in `tesla_comparison_ratios` and apply the percentage difference formula using the corresponding `firm_ratio_standards`. I will store these differences in a new DataFrame called `ratio_differences_percent`.



```python
ratio_differences_percent = pd.DataFrame(index=tesla_comparison_ratios.index)

for ratio in tesla_comparison_ratios.columns:
    firm_std_value = firm_ratio_standards.get(ratio)

    if firm_std_value is not None and firm_std_value != 0:
        # Calculate percentage difference: ((Tesla_Ratio - Firm_Standard) / Firm_Standard) * 100
        difference = ((tesla_comparison_ratios[ratio] - firm_std_value) / firm_std_value) * 100
    else:
        # Handle cases where firm standard is 0 or not found (e.g., set to NaN or 0)
        # For simplicity, if firm standard is 0, we'll just set difference to NaN.
        # If Tesla's ratio is also 0, this might lead to 0/0, which is NaN anyway.
        difference = np.nan

    ratio_differences_percent[ratio] = difference

print("Calculated quarterly percentage differences between Tesla's ratios and firm standards:")
print(ratio_differences_percent.head())
```

**Reasoning**:
To visualize the trends of Tesla's financial ratios relative to firm standards, I will create multi-panel line plots for each ratio, displaying the percentage difference over time. This will clearly show periods of outperformance or underperformance.



```python
fig, axes = plt.subplots(nrows=len(ratio_differences_percent.columns), ncols=1, figsize=(12, 6 * len(ratio_differences_percent.columns)))

for i, ratio in enumerate(ratio_differences_percent.columns):
    sns.lineplot(data=ratio_differences_percent, x=ratio_differences_percent.index, y=ratio, ax=axes[i], marker='o', label=ratio)
    axes[i].set_title(f'Tesla {ratio} vs. Firm Standard (% Difference)')
    axes[i].set_ylabel('% Difference from Standard')
    axes[i].set_xlabel('Date')
    axes[i].axhline(0, color='red', linestyle='--', linewidth=1) # Add a zero line for reference
    axes[i].legend()
    axes[i].grid(True)

plt.tight_layout()
plt.show()
```
